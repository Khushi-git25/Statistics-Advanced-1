Question 1: What is a random variable in probability theory?

In probability theory, a random variable is a numerical outcome of a random phenomenon. More formally:
A random variable is a function that assigns a real number to each outcome in the sample space of a random experiment.

1.	Discrete Random Variable:

A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials of a binary experiment (i.e., an experiment with only two possible outcomes: "success" or "failure").
________________________________________

1.	Fixed number of trials (n)

3.	Probability of success (p) is the same in each trial
4.	Trials are independent

Question 5: What is the standard normal distribution, and why is it important?

Answer:

Definition:
Standard Normal Distribution
The standard normal distribution is a special case of the normal distribution. It is a continuous probability distribution with:
•	Mean (μ) = 0
•	Standard deviation (σ) = 1
It is denoted as:
Z∼N (0,1)
The variable Z is often called a standard normal random variable.
________________________________________
 Key Features:

•	Bell-shaped and symmetric around 0
•	Total area under the curve = 1
•	About:
o	68% of values fall within ±1 standard deviation
o	95% within ±2
o	99.7% within ±3
(This is called the Empirical Rule)
________________________________________
 Why Is It Important?

1.	Basis for Z-scores and Standardization
Any normal distribution can be converted to the standard normal using:
                         Z =   X − μ / σ

This helps compare scores from different normal distributions.
2.	Simplifies Probability Calculations
We can use Z-tables (or software) to find probabilities easily.
3.	Used in Hypothesis Testing
Many statistical tests (e.g., Z-test) are based on the standard normal distribution.
4.	Central Limit Theorem (CLT)
The CLT states that the sampling distribution of the sample mean approaches a normal distribution (often standard normal when standardized), making it crucial in inferential statistics.
________________________________________
 Example:

If test scores are normally distributed with μ=70 and σ=10, what is the standardized score (Z-score) for a student who scored 85?

Z = 85 – 70 / 10 = 1.5

This means the student scored 1.5 standard deviations above the mean.


Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:

 Definition: Central Limit Theorem (CLT)

The Central Limit Theorem (CLT) states that:

When you take a large enough random sample from any population (with a finite mean and variance), the sampling distribution of the sample mean will be approximately normal, regardless of the shape of the original population distribution.
________________________________________
 In Simple Terms:

•	If you repeatedly take samples from a population and calculate the mean of each sample...
•	Then plot the distribution of those sample means...
•	That distribution will look normal as the sample size grows — even if the original data isn't normal.
________________________________________
 Mathematical Form:

If:
•	X1, X2, ..., Xn are independent and identically distributed (i.i.d.) random variables with:
o	Mean: μ
o	Standard deviation: σ
Then the sampling distribution of the sample mean Xˉ
 Xˉ∼ N (μ, σ^2/n)
As n→∞
________________________________________
 Why Is the CLT Critical in Statistics?

1.	Foundation for Inferential Statistics
It allows us to:
o	Construct confidence intervals
o	Perform hypothesis testing
o	Use Z-tests and t-tests, even when the original data is not normally distributed
2.	Works with Non-Normal Populations
As long as the sample size is large enough (usually n≥30n \geq 30), we can make normal-based inferences even from skewed or non-normal populations.
3.	Enables Predictability in Random Processes
The CLT explains why many real-world phenomena (like measurement errors or test scores) tend to be normally distributed when aggregated.

________________________________________
Example:

Suppose the time it takes to assemble a product follows a right-skewed distribution with:
•	Mean μ=20 minutes
•	Standard deviation σ=5
If we take random samples of 40 workers and compute the average assembly time:
•	The distribution of those sample means will be approximately normal, even though the original data is skewed.
•	We can now apply normal-based statistical methods.


Question 7: What is the significance of confidence intervals in statistical analysis?

Answer:

 Definition: Confidence Interval (CI)

A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.

________________________________________
 Significance and Importance of Confidence Intervals
________________________________________
1. Estimates Population Parameters
•	A confidence interval provides a range within which we believe the true population value lies.
•	It's more informative than a single point estimate (like a sample mean).
Example: “We are 95% confident that the average height of adult males is between 175 cm and 182 cm.”
________________________________________
2. Measures Uncertainty
•	It expresses the degree of uncertainty or variability in an estimate.
•	A wider interval means more uncertainty; a narrower interval means more precision.
________________________________________
3. Used in Hypothesis Testing
•	Confidence intervals can be used to test hypotheses:
o	If a hypothesized value (like μ=50\mu = 50) falls outside the interval, it's likely incorrect at that confidence level.
For example, if a 95% CI for the mean is (52, 60), then we can reject the claim that the mean is 50.
________________________________________
4. Helps in Decision-Making
•	In fields like medicine, business, and engineering, CIs help determine if results are statistically and practically significant.
Example: In a clinical trial, if the confidence interval for the difference in recovery times does not include 0, the new treatment is likely effective.
________________________________________
5. Adaptable to Different Confidence Levels
•	Common levels: 90%, 95%, and 99%
•	Higher confidence = wider interval
Lower confidence = narrower interval
________________________________________
 Example:
Suppose a sample of 100 students has an average test score of 75, with a standard deviation of 10.
A 95% confidence interval for the mean is:
75 ± 1.96 × 10/ root over 100 = 75±1.96
⇒(73.04,76.96)
So, we are 95% confident that the true average test score for all students lies between 73.04 and 76.96.
________________________________________
 In Summary:
Feature	Role of Confidence Intervals
Estimation	Estimates the true population parameter
Uncertainty	Shows how precise or uncertain an estimate is
Hypothesis Testing	Helps determine if a claim about a population is valid
Decision Support	Informs evidence-based decisions


Question 8: What is the concept of expected value in a probability distribution?

Answer:

 Definition:
 Expected Value
The expected value (also called the mean or expectation) of a random variable is the long-run average outcome of a random process if it were repeated many times.
In simple terms:
The expected value tells you what you would expect to happen on average.
It’s a weighted average of all possible values that a random variable can take, where the weights are the probabilities of those values.
________________________________________
 1. For a Discrete Random Variable XX:

E(X)= i∑xi⋅P(xi)

Where:

•	xi= possible value of X
•	P(xi) = probability of xi
________________________________________
 2. For a Continuous Random Variable:

E(X)= ∞ ∫−∞ x⋅f (x) dx

Where:
•	f(x)f(x) is the probability density function (PDF)
________________________________________
 Why Is Expected Value Important?

1.	Foundation for Decision Making
o	Used in economics, insurance, gambling, and business to evaluate outcomes.
o	Helps choose between strategies by comparing expected gains or losses.
2.	Summarizes a Probability Distribution

3.	Basis for Variance and Standard Deviation
o	These measures of spread are calculated around the expected value.

Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

Answer:


```python
import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 5
n = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=n)

sample_mean = np.mean(data)
sample_std = np.std(data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")

plt.hist(data, bins=30, edgecolor='black', color='skyblue', density=True)
plt.title("Histogram of Normally Distributed Data (μ=50, σ=5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {sample_mean:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.

● Write the Python code to compute the mean sales and its confidence interval.

Answer:

Retail Sales Analysis Using the Central Limit Theorem (CLT)

You're given daily sales data for 20 days (likely a sample from a larger 2-year dataset). You're tasked with estimating the average daily sales and computing a 95% confidence interval using the Central Limit Theorem (CLT).
________________________________________
 1. How to Apply the Central Limit Theorem (CLT)

To estimate the true average daily sales of the population:

1.	Assume the provided data is a random sample from the larger population of daily sales.
2.	Use the sample mean and sample standard deviation to estimate the population mean.
3.	Apply the CLT:
o	The sampling distribution of the sample mean is approximately normal, since n=20 n is a moderately sized sample.
o	Use the formula for a confidence interval:
                      CI=xˉ±z⋅s/ root over n
           Where:
•	xˉ\bar{x}xˉ = sample mean

•	s = sample standard deviation

•	n = sample size

•	z = z-score for 95% confidence → 1.96


```python
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

n = len(daily_sales)

mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)

z = 1.96

margin_of_error = z * (std_sales / np.sqrt(n))

lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
